```python
import pandas as pd

from google.colab import  drive

drive.mount('/drive')
```

```python
base_credito = pd.read_csv(r"/drive/MyDrive/class_analise_credito.csv")
base_credito.head()
```

```python
print(base_credito.info())
```

```python
#ELIMINANDO LINHAS COM VALORES NULOS NA COLUNA VBEM
base_credito = base_credito.dropna(subset=['VBEM'])

base_credito
```

```python
#ELIMINANDO LINHAS COM VALOR NULO NA COLUNA ESTC

base_credito = base_credito[base_credito['ESTC'] != 0]
base_credito
```

```python
#ENCONTRANDO A MÉDIA DOS VALORES NA COLUNA ESTC

media = base_credito['ESTC'].mean()
print(media)
```

```python
#ELIMINANDO LINHAS COM VALORES 30% MAIORES OU 30% MENORES DO QUE A MÉDIA DA COLUNA

nova_base=base_credito[(base_credito['ESTC']<1.30*media) & (base_credito['ESTC']>0.7*media)]
nova_base
```

```python
#CRIANDO NOVO DATAFRAME DE APENAS 5 COLUNAS

credito_selecionada = base_credito[['ESTC','RENDA','IDADE','NPARC','VBEM']]
credito_selecionada
```

```python
#CRIANDO NOVA COLUNA RELACIONANDO OUTRAS DUAS E NUMEROS ALEATORIOS

import random


credito_selecionada["IDADEXRENDA"] = (credito_selecionada["RENDA"]-credito_selecionada["IDADE"])*random.uniform(0, 1)

credito_selecionada
```

```python
#Embaralhe aleatoriamente o dataframe COM INDICES EMBARALHADOS
shuffled=credito_selecionada.sample(frac=1)

shuffled
```

```python
#Embaralhe aleatoriamente o dataframe COM NOVOS INDICES 

shuffled_c_novos_indices=credito_selecionada.sample(frac=1).reset_index()

shuffled_c_novos_indices
```

```python
#ORDENANDO A BASE DE ACORDO COM A IDADE DO MAIS NOVO PARA O MAIS VELHO
credito_selecionada = credito_selecionada.sort_values(by=['IDADE'], ascending=True)
credito_selecionada
```

```python
#Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da  coluna.
from pandas.core.groupby import groupby

conta_ints = credito_selecionada.groupby('ESTC')[['ESTC']].count()
print(conta_ints)

#TRANSFORMANDO EM STRING A COLUNA VBEM E CONTANDO
credito_selecionada['VBEM'] = credito_selecionada['VBEM'].astype(str)

conta_str= credito_selecionada.groupby('VBEM')[['VBEM']].count()
print(conta_str)
```

```python
#CRIANDO UMA NOVA COLUNA BASEADA NAS OUTRAS NA FORMA STRING
credito_selecionada.loc[credito_selecionada['IDADE'] >=60, "faixa_etaria"] ='idoso'
credito_selecionada.loc[credito_selecionada['IDADE'] >=20, "faixa_etaria"] = 'adulto'
credito_selecionada.loc[credito_selecionada['IDADE']<=19, "faixa_etaria"] = 'adolescente'
credito_selecionada

conta_str= credito_selecionada.groupby('faixa_etaria')[['faixa_etaria']].count()
print(conta_str)
```

```python
#Utilize a função MAPS para fazer algum tipo de substituição no Dataframe
credito_selecionada['ANALISE_ESTC'] = credito_selecionada['ESTC'].map(
    {1: '28 horas de estudo', 2: '56 horas de estudo', 3: '84 horas de estudo'})

credito_selecionada
```

```python
#Exporte o último Dataframe para um arquivo CSV e salve no seu drive
#Exporte o mesmo Dataframe para um arquivo JSON e salve no seu drive
credito_selecionada.to_csv(r'/drive/MyDrive/analise_final_credito.csv')

credito_selecionada.to_json(r'/drive/MyDrive/analise_json_credito.json')
```
